# **Data Cleaning**

```python
import pandas as pd
import numpy as np
from pandas.api.types import is_numeric_dtype
from sklearn.preprocessing import LabelEncoder
from pandas.api.types import is_string_dtype
```

```python
df=pd.read_csv("Data set before cleaning.csv")
```

### data set before Cleaning

```python
df
```

```python
df1=df.copy()
```

```python
df1['The game rviewer'] = df1['The game rviewer'].str.replace('\n',"") #מחיקת תווים מיותרים משם מפרסם הסקירה
```

```python
df1.drop_duplicates(inplace=True) #מחק כפילויות
```

```python
# שינוי תאריך עליית הסקירה מתאריך מלא במלל לשנה
i=0
for date__ in df1['The date of the review']:
    new_date=date__
    v=0
    try:
        v=int(df1['The date of the review'][i].find(",", 0, 20))
    except:
        v=0
        
    if v!=0:
        v=v+2
        new_date=int(date__[v])*1000 + int(date__[v+1])*100 + int(date__[v+2])*10 + int(date__[v+3])
        df1['The date of the review'][i]=float(new_date)
    i+=1

```

```python
# שינוי תאריך עליית המשחק מתאריך מלא במלל לשנה
i=0
for date_ in df1['First Released']:
    new_date=date_
    v=0
    try:
        v=int(df1['First Released'][i].find(",", 0, 20))  
    except:
        v=0
    if v != 0:
        try:
            new_date=int(date_[len(date_)-4])*1000 + int(date_[len(date_)-3])*100 + int(date_[len(date_)-2])*10 + int(date_[len(date_)-1])
            df1['First Released'][i]=float(new_date)
        except:
            df1['First Released'][i]=np.nan
    i+=1
```

```python
#להפוך את הפלטפורה לכמות פלטפורמות שאפשר לשחק את המשחק
i=0
for fit_p in df1['Fit platform']:
    count=fit_p.count(',')
    df1['Fit platform'][i]=float(count+1)
    i+=1
    
```

```python
#מספור השמות 
i=0
for name_ in df1['Name of the game']:
    df1['Name of the game'][i]=i
    i+=1
```

```python
df1 = df1.dropna(subset=['Name of the game']) 
df1 = df1.drop(labels=['Raiting'],axis=1) #להעלים את עמודה Raiting
```

```python
#מניעת תאריכים לא רצוים
w=0
for date_ in df1['First Released']:
    date1=float(df1['First Released'][w])
    if date1>2022:
        df1['First Released'][w]=np.nan
    w+=1    
```

```python
cols=df1[df1.columns[df1.columns!='Grade']]#outliers
for i in cols:
    if is_numeric_dtype(df1[i]):
        z_score=(df1[i]-df1[i].mean())/df1[i].std()
        outliers=(abs(z_score)>3)
        df1[i][outliers]=np.nan
        
    
df1 = df1.dropna()
```

```python
dff=df1.copy()
for i in dff:
    if is_string_dtype(df1[i]):
        if type(df1[i][0])!=float: 
            dff[i]=LabelEncoder().fit_transform(dff[i])
```
